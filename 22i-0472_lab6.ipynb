```python
class depthlimited:
 
    def __init__(self):

        self.vertices = set()
        self.edges = {}


    def add_edge(self, u, v):
        self.vertices.add(u)
        self.vertices.add(v)

        if u not in self.edges:
            self.edges[u] = []

        self.edges[u].append(v)


    def is_valid(self, puzzle):
        
        for i in range(3):
            for j in range(3):
                if puzzle[i][j] == '_':
                    x, y = i, j

        moves = [(x+1, y), (x-1, y), (x, y+1), (x, y-1)]
        valid_moves = [(i, j) for i, j in moves if 0 <= i < 3 and  0 <= j < 3 ]
        return valid_moves, x, y

    
    def dls(self, start, goal, limit=0):

            stack = [(start, 0)]
            visited = set(str(start))

            while stack:

                curr_node, depth = stack.pop()


                if curr_node == goal:
                    print('Found')
                    print(curr_node)
                    return

                # print(curr_node)

                if depth < limit:

                    moves, x, y = self.is_valid(curr_node)

                    for move in moves:

                        new_puzzle = [row.copy() for row in curr_node]
                        new_puzzle[x][y], new_puzzle[move[0]][move[1]] = new_puzzle[move[0]][move[1]], new_puzzle[x][y]
                        str_puzzle = str(new_puzzle)

                        if str_puzzle not in visited:
                            stack.append((new_puzzle, depth+1))
                            visited.add(str_puzzle)

            
            print('Not Found')

            
    def iddfs(self, start, goal):
        limit = 0
        # while limit < 100:
        while True:
            print("Depth limit:", limit)
            if self.dls(start, goal, limit):
                return
            limit += 1

            
g = depthlimited()

start = [['2', '8', '3'], ['1', '6', '4'], ['7', '_', '5']]
goal = [['1', '2', '3'], ['8', '_', '4'], ['7', '6', '5']]

g.dls(start, goal ,  100)
```
